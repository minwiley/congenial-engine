```python
filename = "2015Q1-capitalbikeshare-tripdata.csv"
#number of records in file (excludes header)
n = sum(1 for line in open(filename)) - 1 
#desired sample size
s = 10000 
#the 0-indexed header will not be included in the skip list
skip = sorted(random.sample(range(1,n+1),n-s)) 
df = pd.read_csv(filename, skiprows=skip)
df.to_csv("sample_2015-2016/2015Q1.csv", index=False)

filename = "2015Q2-capitalbikeshare-tripdata.csv"
#number of records in file (excludes header)
n = sum(1 for line in open(filename)) - 1 
#desired sample size
s = 10000 
#the 0-indexed header will not be included in the skip list
skip = sorted(random.sample(range(1,n+1),n-s)) 
df2 = pd.read_csv(filename, skiprows=skip)
df2.to_csv("sample_2015-2016/2015Q2.csv", index=False)

filename = "2015Q3-capitalbikeshare-tripdata.csv"
#number of records in file (excludes header)
n = sum(1 for line in open(filename)) - 1 
#desired sample size
s = 10000 
#the 0-indexed header will not be included in the skip list
skip = sorted(random.sample(range(1,n+1),n-s)) 
df3 = pd.read_csv(filename, skiprows=skip)
df3.to_csv("sample_2015-2016/2015Q3.csv", index=False)

filename = "2015Q4-capitalbikeshare-tripdata.csv"
#number of records in file (excludes header)
n = sum(1 for line in open(filename)) - 1 
#desired sample size
s = 10000 
#the 0-indexed header will not be included in the skip list
skip = sorted(random.sample(range(1,n+1),n-s)) 
df4 = pd.read_csv(filename, skiprows=skip)
df4.to_csv("sample_2015-2016/2015Q4.csv", index=False)
```

```python
path = "sample_2015-2016"
all_files = glob.glob(path + "/*.csv")
frame = pd.DataFrame()
list_15_16 = []

for file_ in all_files:
    df = pd.read_csv(file_,index_col=None, header=0)
    list_15_16.append(df)
frame = pd.concat(list_15_16)
frame.to_csv("sample_2015-2016/sample_2015-2016.csv", index=False)
```
